```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from langchain_groq import ChatGroq
import os
from google.colab import userdata
```

```python
# Get Groq key from Colab secrets (same way you did Replicate)
GROQ_API_KEY = userdata.get('groq_api_key')  # <-- Add this in your Colab secrets

# Optional: Set as environment variable (good practice)
os.environ["GROQ_API_KEY"] = GROQ_API_KEY

# Initialize Groq (replaces Replicate)
output = ChatGroq(
    model="llama-3.1-8b-instant",
    api_key=GROQ_API_KEY,
    temperature=0.7,
)
```

```python
user_query = input ("Ask a question: ")
initial_response = output.invoke(user_query)
print("\nInitial Response (No RAG):\n")
print(initial_response)
```
